```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
```

```python
n = 10000

df = pd.DataFrame({
    "age": np.random.normal(35, 10, n),
    "years_experience": np.random.normal(10, 5, n),
    "salary": np.random.normal(60000, 15000, n),
    "hours_per_week": np.random.normal(40, 5, n),
    "department": np.random.choice([0, 1], n)
})

df.head()
```

```python
df['department'].value_counts()  # how many of each departement type?
```

Step 1 — Basic Understanding

For each column (except department):

Draw a histogram.

Answer:

Is it normally distributed?

Is it skewed?

Are there outliers?

Is the spread wide or narrow?

```python
for col in df.columns[:-1]:
  plt.hist(df[df['department'] == 1][col], color='blue', alpha=0.7, density=True)
  plt.hist(df[df['department'] == 0][col], color='pink', alpha=0.7, density=True)
  plt.title('salary')
  plt.xlabel(col)
  plt.ylabel('Probability')
  plt.legend(['department 1', 'department 0'])
  plt.show()
```
